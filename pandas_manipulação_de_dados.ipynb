#Manipulaçao de dados

```python
import pandas as pd
```

```python
url = "https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_hospedagem.json"
dados_hospedagem = pd.read_json(url)

dados_hospedagem.head()
```

```python
dados_hospedagem = pd.json_normalize(dados_hospedagem['info_moveis'])
```

```python
dados_hospedagem.head(3)
```

```python
colunas = list(dados_hospedagem.columns)
```

```python
dados_hospedagem = dados_hospedagem.explode(colunas[3:])
dados_hospedagem
```

```python
# resetando os indices
dados_hospedagem.reset_index(inplace=True , drop=True)
dados_hospedagem.head()
```

```python
dados_hospedagem.info()
```

```python
# convertendo dados para int e float
import numpy as np
```

```python
dados_hospedagem['max_hospedes'] = dados_hospedagem['max_hospedes'].astype(np.int64)
```

```python
dados_hospedagem.info()
```

```python
columns = ["quantidade_banheiros", "quantidade_quartos", "quantidade_camas"]

dados_hospedagem[columns] = dados_hospedagem[columns].astype(np.int64)

dados_hospedagem.info()
```

```python
dados_hospedagem['avaliacao_geral'] = dados_hospedagem['avaliacao_geral'].astype(np.float64)
```

```python
dados_hospedagem['preco']
```

```python
from os import replace
# retirando $  da coluna preco para poder transformar em float64
# apply , lambda e replace

dados_hospedagem['preco'] = dados_hospedagem['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
dados_hospedagem['preco']
```

```python
dados_hospedagem['preco'] = dados_hospedagem['preco'].astype(np.float64)
```

```python
dados_hospedagem[['taxa_deposito' , 'taxa_limpeza']]
```

```python
dados_hospedagem[['taxa_deposito' , 'taxa_limpeza']] = dados_hospedagem[['taxa_deposito' , 'taxa_limpeza']].applymap(lambda x : x.replace('$','').replace(',','').strip())

dados_hospedagem[['taxa_deposito' , 'taxa_limpeza']]
```

```python
dados_hospedagem[['taxa_deposito' , 'taxa_limpeza']] = dados_hospedagem[['taxa_deposito' , 'taxa_limpeza']].astype(np.float64)

dados_hospedagem.info()
```

```python
# transformando os caracteres em minusculo
dados_hospedagem['descricao_local'] = dados_hospedagem['descricao_local'].str.lower()
```

```python
# removendo caractesres especiais
dados_hospedagem['descricao_local'] = dados_hospedagem['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
# removendo hifens fora de palavras compostas
dados_hospedagem['descricao_local'] = dados_hospedagem['descricao_local'].str.replace('(?<!\w)-(?!\w)', '' , regex=True)
```

```python
dados_hospedagem['descricao_local'][1]
```

```python
# separar caracteres  (Tokenização)
dados_hospedagem['descricao_local'] = dados_hospedagem['descricao_local'].str.split()
```

```python
dados_hospedagem['comodidades']
```

```python
dados_hospedagem['comodidades'] = dados_hospedagem['comodidades'].str.replace('\{|}|\"', '', regex=True)
```

```python
dados_hospedagem['comodidades'] = dados_hospedagem['comodidades'].str.split()

dados_hospedagem['comodidades']
```

```python
# coluna descricao_vizinhanca

dados_hospedagem['descricao_vizinhanca'] = dados_hospedagem['descricao_vizinhanca'].str.lower()
```

```python
dados_hospedagem['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', " " , regex=True)
```

```python
dados_hospedagem['descricao_vizinhanca'] = dados_hospedagem['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)' , "" , regex=True)
```

```python
dados_hospedagem['descricao_vizinhanca'] = dados_hospedagem['descricao_vizinhanca'].str.split()
```

```python
url = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json'

data = pd.read_json(url)

data
```

```python
data = pd.json_normalize(data['dados_locacao'])

data
```

```python
# retirando as listas
colunas = list(data.columns)

# output : ['apartamento',
#  'datas_combinadas_pagamento',
#  'datas_de_pagamento',
#  'valor_aluguel']

data = data.explode(colunas[1:])

data
```

```python
data.reset_index(drop=True,inplace=True)
```

```python
data.info()
```

```python
data['valor_aluguel'].head(5)
```

```python
data['valor_aluguel'] = data['valor_aluguel'].apply(lambda x : x.replace("$" , "").replace("," , "").replace("reais" , "").strip())

data['valor_aluguel'].head(5)
```

```python
data['valor_aluguel'] = data['valor_aluguel'].astype(np.float64)

data.info()
```

```python
data.head(5)
```

```python
# removendo blocoAP
data['apartamento'] = data['apartamento'].str.replace('\(blocoAP\)' , '' , regex=True)
```

```python
data.info()
```

```python
data['datas_de_pagamento'] = pd.to_datetime(data['datas_de_pagamento'], format='%d/%m/%Y')
data['datas_combinadas_pagamento'] = pd.to_datetime(data['datas_combinadas_pagamento'], format='%d/%m/%Y')
```
